```python
p_df = panel_df[['P99AG', 'P27FV01']]
```

```python
p_df.rename(columns={'P99AG':'Age', 'P27FV01':'M_values'}, inplace=True)
```

```python
p_df.info()
```

```python
p_df.boxplot()
```

```python
for idx, i in enumerate(p_df['M_values']):
    if i<0:
        p_df['M_values'][idx] = None
```

```python
ls = [5,6,7,8,9]
for idx, i in enumerate(ls):
    print("인덱스 넘버는 ", idx)
    print("값은 ", i)
```

```python
p_df = p_df.dropna(axis=0)
```

```python
p_df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = p_df['Age']
Y = p_df['M_values']
```

```python
model = LinearRegression()
model.fit(X.values.reshape(-1,1), Y)
```

```python
# 1 매우 그렇다
# 2 조금 그렇다
# 3 별로 그렇지 않다
# 4 전혀 그렇지 않다
```

```python
model.predict([[20]])
```

```python
model.predict([[50]])
```

```python
model.predict([[80]])
```

```python
plt.plot(X, Y, 'o')
plt.plot(X, model.predict(X.values.reshape(-1,1)))
```

```python
test_list = [[], []]
for age in range(18, 86):
    test_list[0].append(age)
    test_list[1].append(model.predict([[age]])[0])
```
